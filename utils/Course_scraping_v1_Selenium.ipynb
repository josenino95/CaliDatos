```python
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# setup chrome options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') # ensure GUI is off
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

# set path to chromedriver as per your configuration
chromedriver_autoinstaller.install()

# set up the webdriver
# driver.quit()
driver = webdriver.Chrome(options=chrome_options)
```

```python
driver.get("https://catalog.ucsb.edu/courses?cq=&page=1")
# driver.quit()
```

```python
wait = WebDriverWait(driver, 10)
clickable_element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.course-button')))
clickable_element.click()
text_element = driver.find_elements(By.CSS_SELECTOR, 'div.field-value')
for i in text_element:
  print(i.text)
driver.quit()
```

```python
def scrape_courses(driver, url):
    driver.get(url)
    time.sleep(5)  # Wait for dynamic content to load

    codes = driver.find_elements(By.CSS_SELECTOR, 'h6[data-v-70adec07]')
    courses = driver.find_elements(By.CSS_SELECTOR, 'h3[data-v-70adec07]')
    descriptions = driver.find_elements(By.CSS_SELECTOR, 'p[data-v-70adec07]')
    descriptions2 = driver.find_elements(By.CSS_SELECTOR, 'p[data-v-70adec07]')

    course_info = [(code.text, course.text, desc.text) for code, course, desc in zip(codes, courses, descriptions)]
    return course_info
```

```python
base_url = "https://catalog.ucsb.edu/courses?cq=&page="
all_courses = []

# Adjust the range for the number of pages you want to scrape
for page in range(1, 2):  # Example: first 5 pages
    url = base_url + str(page)
    all_courses.extend(scrape_courses(driver, url))

driver.quit()

# Create a DataFrame and save to Excel
df = pd.DataFrame(all_courses, columns=['Code', 'Course Name', 'Description'])
```
